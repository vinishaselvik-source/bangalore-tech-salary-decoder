```python
#load and inspect
import pandas as pd
df=pd.read_csv('bangalore_tech_salaries.csv')

print(df)
```

```python
print(df.info())
print("Shape:", df.shape)
print("datatype",df.dtypes)
print("missing value:",df.isnull().sum())
print("duplicates:",df.duplicated().sum())
print("summary:",df.describe(include='all'))
```

```python
# TASK 2 : Clean

import pandas as pd
import numpy as np

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")


df.drop_duplicates(inplace=True)


role_map = {
    'sde': 'Software Engineer',
    'software engg': 'Software Engineer',
    'software developer': 'Software Engineer',
    'ds': 'Data Scientist',
    'devops': 'DevOps Engineer',
    'product lead': 'Product Lead',
    'ba': 'Business Analyst',
    'da': 'Data Analyst',
    'pm': 'Product Manager',
    'hr': 'Human Resources',
    'qa': 'Quality Assurance',
    'ui/ux': 'UI/UX Designer',
    'sde backend': 'Software Engineer',
    'sde frontend': 'Software Engineer',
    'software engineer': 'Software Engineer',
    'ml engineer': 'ML Engineer',
    'data scientist': 'Data Scientist',
    'data analyst': 'Data Analyst',
    'product manager': 'Product Manager',
    'product designer': 'Product Designer',
    'site reliability engineer': 'Site Reliability Engineer',
    'analytics engineer': 'Analytics Engineer',
    'bi analyst': 'BI Analyst',
    'backend dev': 'Backend Developer',
    'backend developer': 'Backend Developer',
    'backend engineer': 'Backend Engineer',
    'business systems analyst': 'Business Systems Analyst',
    'designer': 'Designer',
    'devops engineer': 'DevOps Engineer',
    'fe': 'Frontend Engineer',
    'frontend dev': 'Frontend Developer',
    'frontend developer': 'Frontend Developer',
    'frontend engineer': 'Frontend Engineer',
    'full stack engineer': 'Full Stack Engineer',
    'fullstack': 'Full Stack Engineer',
    'fullstack dev': 'Full Stack Engineer',
    'infra engineer': 'Infra Engineer',
    'sde fs': 'Software Engineer',
    'sde full-stack': 'Software Engineer',
    'sde-backend': 'Software Engineer',
    'sde-frontend': 'Software Engineer',
    'sre': 'Site Reliability Engineer',
    'sr pm': 'Product Manager',
    'ui designer': 'UI Designer',
    'ux designer': 'UX Designer',

    'business analyst': 'Business Analyst',
    'be': 'Backend Engineer',
    'data science engineer': 'Data Scientist'
}
education_map = {
    'tier1': 'Tier 1',
    'tier-1': 'Tier 1',
    '1': 'Tier 1',
    't1': 'Tier 1',
    'tier2': 'Tier 2',
    'tier-2': 'Tier 2',
    '2': 'Tier 2',
    't2': 'Tier 2',
    'tier3': 'Tier 3',
    'tier-3': 'Tier 3',
    '3': 'Tier 3',
    't3': 'Tier 3'
}


company_map = {
    "product based": "Product",
    "product": "Product",
    "service based": "Service",
    "service": "Service",
    "startup": "Startup",
    "start-up": "Startup",
    "mnc": "MNC",
    "mid-size": "Mid-size",
    "unicorn": "Unicorn",
    "early-stage": "Early-stage",
    "MID-SIZE": "Mid-size",
    "early stage": "Early-stage"
}


df["role_clean"] = df["role"].astype(str).str.lower().str.strip()
df["education_tier_clean"] = df["education_tier"].astype(str).str.lower().str.strip()
df["company_type_clean"] = df["company_type"].astype(str).str.lower().str.strip()

df["role_clean"] = df["role_clean"].replace(role_map)
df["education_tier_clean"] = df["education_tier_clean"].replace(education_map)
df["company_type_clean"] = df["company_type_clean"].replace(company_map)

known_roles = set(role_map.values())
df.loc[~df["role_clean"].isin(known_roles), "role_clean"] = 'Unknown'

known_education_tiers = set(education_map.values())
df.loc[~df["education_tier_clean"].isin(known_education_tiers), "education_tier_clean"] = 'Unknown'

known_company_types = set(company_map.values())
df.loc[~df["company_type_clean"].isin(known_company_types), "company_type_clean"] = 'Unknown'
df['role_clean'] = df['role_clean'].fillna('Unknown')
df['company_type_clean'] = df['company_type_clean'].fillna('Unknown')
df['education_tier_clean'] = df['education_tier_clean'].fillna('Unknown')
def convert_ctc(x):
    if pd.isna(x):
        return np.nan
    x_str = str(x).lower()
    x_str_numeric = x_str.replace("₹", "").replace("$", "").replace(",", "").replace("lpa", "").strip()

    try:
        value = float(x_str_numeric)
        if 'lpa' in x_str:
            return value
        elif value > 1000:
            return value / 100000
        else:
            return value
    except ValueError:
        return np.nan
df["current_ctc"] = df["current_ctc"].apply(convert_ctc)
print(f"NaNs in current_ctc after conversion (before imputation): {df['current_ctc'].isna().sum()}")
df["previous_ctc"] = df["previous_ctc"].apply(convert_ctc)
print(f"NaNs in previous_ctc after conversion (before imputation): {df['previous_ctc'].isna().sum()}")
df['current_ctc'] = df['current_ctc'].fillna(df['current_ctc'].median())
df['previous_ctc'] = df['previous_ctc'].fillna(df['previous_ctc'].median())
df["median_ctc"] = df.groupby("role_clean")["current_ctc"].transform("median")
df["gap"] = df["median_ctc"] - df["current_ctc"]
print("DataFrame dtypes after cleaning:")
print(df.dtypes)
print("\nRole Counts after cleaning:")
print(df["role_clean"].value_counts(dropna=False))
print("\nCompany Type Counts after cleaning:")
print(df["company_type_clean"].value_counts(dropna=False))
print("\nEducation Tier Counts after cleaning:")
print(df["education_tier_clean"].value_counts(dropna=False))
df.to_csv("further_cleaned_bangalore_tech_salaries.csv", index=False)
print("\nFirst 5 rows after cleaning:")
print(df.head())
print("\nInfo after cleaning:")
df.info()
```
